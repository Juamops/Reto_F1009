## Visualizacion de datos con reduccion de dimensiones usando PCA

Para este NoteBook ajustaremos un modelo de PCA para visualizar los datos n-dimesionales en una grafica de 2 dimensiones que mantenga la mayor cantidad de informacion sobre los datos

```python
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from umap import UMAP
```

```python
df = pd.read_csv('../data/processed_data.csv')
df = df.drop('Unnamed: 0', axis=1)
```

```python
print(df.columns)
```

```python
df.describe()
```

```python
selected_cols = ['Gender', 'Age', 'Sleep Duration', 
                 'Stress Level', 'BMI Category', 'Heart Rate', 'Sleep Disorder']

X = df[selected_cols]
pca = PCA(n_components=2)
pca.fit(X)
X_trans = pca.transform(X)
```

```python
X_trans_pca = X_trans.transpose()
X_trans_pca.shape
```

```python
normalized_quality = (df['Quality of Sleep'] - df['Quality of Sleep'].min()) / df['Quality of Sleep'].max()

plt.scatter(X_trans_pca[0], X_trans_pca[1], c=normalized_quality)
plt.magma()
plt.show()
```

```python
pca_q_df = pd.DataFrame({'PCA1': X_trans_pca[0], 'PCA2': X_trans_pca[1], 'Quality': df['Quality of Sleep']-4, 'norm_q': normalized_quality})
pca_q_df['Quality'].value_counts()
```

```python
for i in range(3):
    tmp_df = pca_q_df[(pca_q_df['Quality'] == 2*i) | (pca_q_df['Quality'] == 2*i + 1)]
    plt.scatter(tmp_df['PCA1'], tmp_df['PCA2'])
    plt.show()
```

```python
tsne = TSNE(n_components=2)
tsne.fit(X)
X_trans_tsne = tsne.fit_transform(X)
X_trans_tsne = X_trans_tsne.transpose()
```

```python
plt.scatter(X_trans_tsne[0], X_trans_tsne[1], c=normalized_quality)
plt.magma()
plt.show()
```
