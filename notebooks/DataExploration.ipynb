## Columnas categoricas

Primero podemos explorar las columnas categoricas para ver cuales son las categorias que estan presentes. Con esto podemos ver si vale la pena comabiar las categorias a valores numericos

```python
print(df['Gender'].unique())
print(df['Occupation'].unique())
print(df['BMI Category'].unique())
print(df['Sleep Disorder'].unique())
```

```python
df['Occupation'].value_counts()
```

```python
# Entramos discrepancia entre lo que dice en la documentacion de la base de datos y lo que esta. 
# Tenemos "Normal" y "Normal Weight" y nos falta "Underweight", sospechamos que hubo un error y vamos
# a corregirlo.

df['BMI Category'].value_counts()
```

```python
df['BMI Category'] = df['BMI Category'].replace({'Normal Weight':'Underweight'})
print(df['BMI Category'].unique())
```

```python
# Ahora cambiaremos a valores nunmericos
df['Gender'] = df['Gender'].replace({'Male':1, 'Female':0})
df['BMI Category'] = df['BMI Category'].replace({'Underweight': 0, 'Normal': 1, 'Overweight': 2, 'Obese': 3 })
df['Sleep Disorder'] = df['Sleep Disorder'].replace({np.NaN:0, 'Sleep Apnea':1, 'Insomnia':1})
```

```python
df.info()
```

```python
df[['Systolic Pressure', 'Diastolic Pressure']] = df['Blood Pressure'].str.split('/', expand=True).astype('int')
```
