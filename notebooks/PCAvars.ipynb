## Use PCA to determine the most important features

The vector of most importance will be used to determine the relative importance of the indiviudal features.

```python
import pandas as pd
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
import numpy as np
```

```python
df = pd.read_csv('../data/processed_data.csv')
df = df.drop('Unnamed: 0', axis=1)
df
```

```python
selected_cols = ['Gender', 'Age', 'Sleep Duration', 
                 'Physical Activity Level', 
                 'Stress Level','BMI Category', 
                 'Heart Rate', 'Daily Steps','Sleep Disorder', 
                 'Systolic Pressure', 'Diastolic Pressure']

# selected_cols = ['Gender', 'Age', 'Sleep Duration', 
#                  'Stress Level', 'BMI Category', 'Heart Rate', 'Sleep Disorder']

X = df[selected_cols]

scaler = StandardScaler()
data_scaled = scaler.fit_transform(X)

pca = PCA(n_components=len(selected_cols))
pca.fit(data_scaled)

vecs = pca.components_
e_vals = pca.explained_variance_

abs_mult_vecs = np.abs(vecs * e_vals)

for i, n in zip(np.argsort(abs_mult_vecs[:2].reshape(-1)), np.sort(abs_mult_vecs[:2].reshape(-1))):
    print(f"{selected_cols[i%len(selected_cols)]} : {' ' * (25-len(selected_cols[i%len(selected_cols)]))}{n}")
```
